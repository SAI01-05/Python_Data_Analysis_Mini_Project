# I-PHONE SALES ANALYSIS

```python
# Importing libraries which are required
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Reading a csv data files
data=pd.read_csv("Apple_Products_Dataset.csv")
data
```

```python
# Checking the null values in the data 
data.isnull().sum()
```

```python
# Describing the data only for integer value
data.describe()
```

# I PHONE SALES ANALYSIS IN INDIA

```python
# [For full information about product with sorting of star_rating]
data.sort_values(["Star Rating"] ,ascending=False).head() 
```

```python
# [Sort and only display the particular column]
highest_Rated_Data=data.sort_values(["Star Rating"] ,ascending=False).head(10)
print(highest_Rated_Data["Product Name"])
```

# GRAPH CHART FOR NUMBER OF RATINGS FOR HIGHEST RATED I-PHONES

```python
# Counting the occurrences of each iPhone model name
iphones = highest_Rated_Data["Product Name"].value_counts()

# Getting the iPhone model names as labels for x-axis
labels = iphones.index

# Getting the number of ratings for each iPhone as y-axis values
counts = highest_Rated_Data["Number Of Ratings"]

# Creating a bar chart using plotly express
# x = iPhone model names, y = number of ratings
figure = px.bar(highest_Rated_Data, x=labels, y=counts, title="Numbers of ratings of highest rated I-Phones")

# Displaying the chart
figure.show()
```

```python
# Counting the occurrences of each iPhone model name
iphones = highest_Rated_Data["Product Name"].value_counts()

# Getting the iPhone model names as labels for x-axis
labels = iphones.index

# Getting the number of reviews for each iPhone as y-axis values
counts = highest_Rated_Data["Number Of Reviews"]

# Creating a bar chart using plotly express
# x = iPhone model names, y = number of reviews
figure = px.bar(highest_Rated_Data, x=labels, y=counts, title="Numbers of Reviews of highest rated I-Phones")

# Displaying the chart
figure.show()
```

```python
# Creating a scatter plot to show relationship between number of ratings and sale price
# x = Number of Ratings, y = Sale Price
# size = bubble size based on Discount Percentage
# trendline = "ols" adds a straight line showing the overall trend
figure = px.scatter(data_frame=data, x="Number Of Ratings", y="Sale Price", size="Discount Percentage", trendline="ols",
                    title="Relationship between sale price and number of rating")

# Displaying the chart
figure.show()
```

```python
# Creating a scatter plot to show relationship between number of ratings and discount percentage
# x = Number of Ratings, y = Discount Percentage
# size = bubble size based on Sale Price
# trendline = "ols" adds a straight line showing the overall trend
figure = px.scatter(data_frame=data, x="Number Of Ratings", y="Discount Percentage", size="Sale Price", trendline="ols",
                    title="Relationship between discount percentage and Number of rating")

# Displaying the chart
figure.show()
```
